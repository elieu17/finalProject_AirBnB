```python
#function for scraping the NYPD website for zipcode:precinct data
def NYPD_scraper(zipcode):
    #browser = init_browser()
    #driver = webdriver.Chrome()
    #url = "https://www1.nyc.gov/site/nypd/bureaus/patrol/find-your-precinct.page"
    #driver.get(url)
    iframe = driver.find_elements_by_tag_name('iframe')[0]
    driver.switch_to.frame(iframe)
    inputElement = driver.find_element_by_xpath("//div[2]/div/form/div/input")
    inputElement.send_keys(zipcode, Keys.ENTER)
    driver.switch_to.default_content()
    time.sleep(2)
    precinct = driver.find_element_by_id('sectorInfo').find_element_by_tag_name('h1').text
    time.sleep(2)
    driver.refresh()
    return precinct
```

```python
#TEST 2
testSet = ['10001','10002','10003']
zipPrct_dict = {}
driver = webdriver.Chrome()
url = "https://www1.nyc.gov/site/nypd/bureaus/patrol/find-your-precinct.page"
driver.get(url)

for zipcode in testSet:
    prct = NYPD_scraper(zipcode)
    newEntry = {zipcode:prct}
    zipPrct_dict.update(newEntry)
    
zipPrct_df = pd.DataFrame.from_dict(zipPrct_dict.items())
zipPrct_df = zipPrct_df.rename(index=int, columns={0: "Zipcode", 1: "Precinct"})
zipPrct_df.to_csv("Zipcodes and Precincts_Test.csv", sep=',')
```

```python
#211 distinct zipcodes/ 1-55
zipcodes1to55 = distinct_zipcodes[1:56]
zipcodes56to111 = distinct_zipcodes[57:112]
zipcodes113to168 = distinct_zipcodes[113:169]
zipcodes169toEnd = distinct_zipcodes[169:]
```

```python
PrctZipDict1to55 = {}
driver = webdriver.Chrome()
url = "https://www1.nyc.gov/site/nypd/bureaus/patrol/find-your-precinct.page"
driver.get(url)
for zipcode in zipcodes1to55:
    try:
        prct = NYPD_scraper(zipcode)
        newEntry = {zipcode:prct}
        PrctZipDict1to55.update(newEntry)
    except: 
        driver.refresh()
    
df_1to55 = pd.DataFrame.from_dict(PrctZipDict1to55.items())
df_1to55 = df_1to55.rename(index=int, columns={0: "Zipcode", 1: "Precinct"})
df_1to55.to_csv("Zipcodes and Precincts_1-55.csv", sep=',')
```
